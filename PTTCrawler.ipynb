```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import time
import os
options= Options()
options.add_argument('--headless')
options.add_argument('--disable-gpu')
driver=webdriver.Chrome(os.getcwd()+"/chromedriver",chrome_options=options)
f=open(os.getcwd()+"/data.txt",'w',encoding='utf8')
import pymysql.cursors
connection=pymysql.connect(host='localhost',
user='root',
password='27727302',
db='example',
charset='utf8mb4',
cursorclass=pymysql.cursors.DictCursor)
for i in range(1,3):
    pagenumber=str(i)
    driver.get('https://www.ptt.cc/bbs/LoL/index'+pagenumber+'.html')
    time.sleep(0.5)
    sourceCode=BeautifulSoup(driver.page_source,"html.parser")
    articles=sourceCode.find_all("div","r-ent")
    for article in articles:
        sourceCode2=BeautifulSoup(str(article),"html.parser")
        title=sourceCode2.find_all("div","title")[0].text
        title=title.strip()
        if title.startswith("[公告]"):
            pass
        else:
            date=sourceCode2.find_all("div","date")[0].text
            number=sourceCode2.find_all("div","nrec")[0].text
            author=sourceCode2.find_all("div","author")[0].text
            print(title.strip())
            print(author)
            print(date)
            number=number.replace("爆",">100")
            print(number)
            data=title+"\n"+number+"\n"+author+"\n"+date
            f.write(data)
f.close()
```
